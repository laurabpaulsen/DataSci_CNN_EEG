```python
import mne
from pathlib import Path
```

```python
# load data
data_path = Path('data')
raw = mne.io.read_raw_fif(data_path / "raw_lau" / "sub-01" / "ses-meg" / "meg" / "oddball_absence-tsss-mc_meg-1.fif")
```

```python


# 
picks = mne.pick_types(raw.info, meg="mag", eeg=False, eog=False, stim=False)

# pick only motor channels
raw.pick(picks)

# plot the sensors
raw.plot_sensors(show_names=True);

# how many channels do we have?
print(raw.info)
```
